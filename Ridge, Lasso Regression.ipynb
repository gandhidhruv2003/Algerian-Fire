```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Algerian_forest_fires_dataset.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip()
```

```python
df.info()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:123,"Region"] = 0
df.loc[123:,"Region"] = 1
```

```python
df
```

```python
df.info()
```

```python
df["Classes"] = np.where(df["Classes"].str.contains("not fire"), "not fire", "fire")
```

```python
df["Region"] = df["Region"].astype(int)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df["day"].fillna(df["day"].mode(), inplace=True)
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.info()
```

```python
df[["day", "month", "year", "Temperature", "RH", "Ws"]] = df[["day", "month", "year", "Temperature", "RH", "Ws"]].astype(int)
```

```python
df[["Rain", "FFMC", "DMC", "DC", "ISI", "BUI", "FWI"]] = df[["Rain", "FFMC", "DMC", "DC", "ISI", "BUI", "FWI"]].astype(float)
```

```python
df.to_csv("Algerian_forest_fires_cleaned_dataset.csv")
```

```python
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
df_copy["Classes"] = np.where(df_copy["Classes"].str.contains("not fire"), 0, 1)
```

```python
df_copy["Classes"].value_counts()
```

```python
plt.style.use("seaborn")
df_copy.hist(bins=50, figsize=(20, 12))
plt.show()
```

```python
percentage = df_copy["Classes"].value_counts(normalize=True)*100

classlabel = ["Fire", "Not Fire"]
plt.figure(figsize=(12, 7))
plt.pie(percentage,labels= classlabel)
plt.show()
```

```python
df_copy.corr()
```

```python
sns.heatmap(df_copy.corr())
plt.show()
```

```python
sns.boxplot(df_copy["FWI"])
plt.show()
```

```python
df_temp = df.loc[df["Region"] == 1]
plt.subplots(figsize=(13, 7))
sns.set_style("whitegrid")
sns.countplot(x="month", hue="Classes", data=df_temp)
plt.ylabel("Number of fire")
plt.title("Fire Analysis for Sidi-Bel Region")
plt.show()
```

```python
df_temp = df.loc[df["Region"] == 0]
plt.subplots(figsize=(13, 7))
sns.set_style("whitegrid")
sns.countplot(x="month", hue="Classes", data=df_temp)
plt.ylabel("Number of fire")
plt.title("Fire Analysis for Bejaia Region")
plt.show()
```
